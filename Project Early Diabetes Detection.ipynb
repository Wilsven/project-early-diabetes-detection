## Part 1: Data Collection and Cleaning
1. Collect diabetes data from UCI Repository

```python
# import pnadas library
import pandas as pd
```

```python
# check dtypes of columns
df.dtypes
```

```python
# assign new lowercase columns to df
df.columns = df.columns.str.lower()
df
```

### Is obesity related to diabetes status?

```python
obesity_diabetes_crosstab = pd.crosstab(df["class"], df["obesity"])
obesity_diabetes_crosstab
```

Values from the crosstab deviates quite significantly from the indpendent values in the array. Therefore, we can say there is a relationship between gender and diabetes. Another thing to note is that the p value is very low (i.e. 3.29 x 10^-24), indicating that the relationship is significant.

### Is there a relationship between age and diabetic status?

```python
sns.boxplot(df["class"], df["age"])
```

```python
no_diabetes = df[df["class"] == 0]
no_diabetes 
```

```python
# let's take a look at the mean of the ages of people without diabetes
no_diabetes["age"].mean()
```

```python
# let's take a look at the median of the ages of people without diabetes
no_diabetes["age"].median()
```

```python
# do the same for people who are diabetic 
diabetes = df[df["class"] == 1]

# mean of the ages of people who are diabetic
print("The mean age of the diabetic is: {}".format(diabetes["age"].mean()))

# median of the ages of people who are diabetic 
print("The median age of the diabetic is: {}".format(diabetes["age"].median()))
```

### Check if the age variable is normally distributed with a qqplot

```python
qqplot(df["age"], fit=True, line="s")
plt.show() 
```

We can see that most points fall on the line with slight skews on the ends of the plot. Therefore, for most parts, we can conclude that age is mostly normally distributed.

So, we can use statistical tests suited for normally distributed data, and in our case, we can conduct a z-test. 

```python
# conduct z test of difference
results = ztest(diabetes["age"], no_diabetes["age"])
print(results)

print("z-score: {}".format(results[0]))
print("p-value: {}".format(results[1]))
```

Since the p-value is less than 0.05, we can reject the null hypothesis and conclude that the differences in ages are significant. 

**Note:** The p-value is the probability of seeing such a result, given that the null hypothesis is true. Therefore, if the null hypothesis were true and we are able to observe a result of probability less than 0.05, we must conclude thay the results are significant and reject the null hypothesis and accept the alternative hypothesis.

In this context, the z-test tells us that older people tend to have a higher chance of being diabetic. By comparing the results above obtained from diabetic and non-diabetic people, the median ages were 48.0 and 45.0 respectively.

While the difference is small, it is nontheless a significant difference as concluded from our z-test.

## Part 3: Machine Learning Model Training
1. Train a baseline model using DummyClassifier
2. Train three different models – LogisticRegression, DescisionTreeClassifier, RandomForestClassifier
3. Identify important features in the best performing model

```python
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 

from sklearn.metrics import confusion_matrix, classification_report
```

```python
# prepare indpendent and dependent variables
pd.read_csv("diabetes_data_cleaned.csv")

X = df.drop("class", axis=1)
y = df["class"]
```

```python
# split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
# begin model training
# start with DummyClassifier to establish baseline
dummy = DummyClassifier(strategy="stratified")
dummy.fit(X_train, y_train)
dummy_pred = dummy.predict(X_test)
```

```python
# access DummyClassifier model
result = confusion_matrix(y_test, dummy_pred)
result
```

```python
print("True Negatives: {}".format(result[0][0]))
print("True Positives: {}".format(result[1][1]))
print("False Negatives: {}".format(result[1][0]))
print("False Positives: {}".format(result[0][1]))
```

We can see that the model does not do a very good job at predicting. As you can see, 21 patients **do not** have diabetes but are yet predicted to be diabetic by the model. This means the patients would potentially have to undergo various tests and additional anxiety.

Similarly, there are 30 patients who are diabetic but are predicted to be non-diabetic by the model. This is bad. What are the implications? This means that the disease is left undiagnosed. Therefore, if we were to completely rely on this machine learning model, we would have missed a substantial number of diabetic patients.

## Question
Which is worse? False positives or false negatives? In the context of healthcare, there are certain considerations to consider.

```python
# use a classification report
print(classification_report(y_test, dummy_pred))
```

## Now, let's try with other machine learning models

```python
# start with LogisticRegression
logr = LogisticRegression(max_iter=10000)
logr.fit(X_train, y_train)
logr_pred = logr.predict(X_test)
```

```python
confusion_matrix(y_test, logr_pred)
```

```python
print(classification_report(y_test, logr_pred))
```

As you can see, there is an improvement in this LogisticRegression model compared to our DummyClassifier model. A 93.0% vs 48.0%. 

```python
# try DecisionTree
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
tree_pred = tree.predict(X_test)
```

```python
confusion_matrix(y_test, tree_pred)
```

```python
print(classification_report(y_test, tree_pred))
```

```python
# try RandomForest
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
forest_pred = forest.predict(X_test)
```

```python
confusion_matrix(y_test, forest_pred)
```

```python
print(classification_report(y_test, forest_pred))
```

We can see that the RandomForest model performed the best against other models and gave the best predicitions. So why is the RandomForestClassifier so good? What are the features that allow it to predict with such high accuracy?

```python
forest.feature_importances_
```

```python
X.columns
```

```python
# build a dataframe to help us visualize the importance of each feature
pd.DataFrame({"feature": X.columns,
             "importance": forest.feature_importances_})
```

We can see from the dataframe that polyuria and polydipsia are both good predictors of diabetes status.

```python
# sort dataframe with reagrds to its importance in descending order for easy reference
pd.DataFrame({"feature": X.columns,
             "importance": forest.feature_importances_}).sort_values("importance", ascending=False)
```
